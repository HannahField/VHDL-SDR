```julia
using GLMakie
```

```julia
levels(M) = collect(-(sqrt(M)-1):2:(sqrt(M)-1))
```

```julia
levels(64)
```

```julia
MQAM(M) = vcat(map(i -> map(j -> levels(M)[i]+1im*levels(M)[j],eachindex(levels(M))),eachindex(levels(M)))...)
MPSK(M) = map(i -> exp(2im*pi*i/M),0:(M-1))
MPAM(M) = collect((-M/2-1):2:(M/2+1))
```

```julia
plot_mod(points,ax,color) = scatter!(ax,real.(points),imag.(points),markersize=8,color=color)
```

```julia
fig = Figure()
ax = Axis(fig[1,1],xlabel="In-phase",ylabel="Quadrature",title="Constellation Diagram",
    aspect=AxisAspect(1),xticks=-3:1:3,yticks=-3:1:3)
QAM16 = plot_mod(MQAM(16),ax,:black)
PSK16 = plot_mod(MPSK(16),ax,:green)
PAM4 = plot_mod(MPAM(4),ax,:red)
Legend(fig[1,2],[QAM16,PSK16,PAM4],["16QAM","16PSK","4PAM"])
fig
```
