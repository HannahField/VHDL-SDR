```julia
function MQAM(M)
    L = Int(sqrt(M))
    levels = collect(-(L-1):2:(L-1))
    avg_pwr = 2*(L^2-1)/3
    levels = round.(Int32,levels/sqrt(avg_pwr)*((2^22)-1))
    return levels
end
```

```julia
function to_quartus(x)
    for i in eachindex(x)
        print("to_signed(" * string(x[i]) * ", 24), ")
        if (i % 4 == 0)
            println()
        end
    end
end
```
